```python
to_do_list=[]
def add_task():
  task=input("enter the task:")
  to_do_list.append({"Task":task,"Status":"pending"})
  print("task added successfully")
def view_task():
  print("To do list is.....")
  if len(to_do_list)==0:
    print("to do list is empty")
  else:
    for index,task in enumerate(to_do_list,1):
      print(f"{index}:{task}")
def remove_task():
  if len(to_do_list)==0:
    print("to do list is empty")
  else:
    task_index=int(input("enter the task index to remove:"))
    if 1<=task_index<=len(to_do_list):
      removed_task=to_do_list.pop(task_index-1)
      print(f"task {removed_task['Task']} removed")
    else:
      print("invalid task index")
def mark_task():
  if len(to_do_list)==0:
    print("to do list is empty")
  else:
    task_index=int(input("enter the task index to mark as done:"))
    if 1<=task_index<=len(to_do_list):
      to_do_list[task_index-1]["Status"]="done"
      print(f"task {to_do_list[task_index-1]['Task']} marked as done")
    else:
      print("invalid task index")
def menu():
  while(True):
    print("------select a option------")
    print("1.Add a new task")
    print("2.view all tasks")
    print("3.remove task")
    print("4.mark task as done")
    print("5.exit")
    choice=int(input("enter your option:"))
    if choice==1:
      add_task()
    elif choice==2:
      view_task()
    elif choice==3:
      remove_task()
    elif choice==4:
      mark_task()
    elif choice==5:
      print("exiting......")
      break
    else:
      print("invalid option")
menu()



```
